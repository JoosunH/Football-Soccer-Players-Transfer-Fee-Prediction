<h3>Created a function and remove the columns if its below threshold.<h3>


```python
def correlation(dataset, threshold):
    col_corr = set()
    #Correlation of dataset only numeric value.
    corr_matrix = dataset.corr(numeric_only=True)
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            #Making a i times J square and storing it.
            if abs(corr_matrix.iloc[i, j]) > threshold:
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
          
```

```python
#Check how many features got removed.
corr_features = correlation(X, 0.70)
len(set(corr_features))
```

<h2>Now do preprocessing and scalization using OneHotEncoder, seperate the value to categorical and numerical values.<h2>

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
# Create the cat and num columns
# Get a list of column names from the 'X' DataFrame that are of numerical data types.
# Get a list of column names from the 'X' DataFrame that are not of numerical data types.

num_cols = X.select_dtypes(include='number').columns.to_list()
cat_cols = X.select_dtypes(exclude='number').columns.to_list()


# Create pipelines for numeric and categorical columns
num_pipeline = make_pipeline(SimpleImputer(strategy='mean'), StandardScaler())
cat_pipeline = make_pipeline(SimpleImputer(strategy='most_frequent'), OneHotEncoder(sparse_output=False))

# Use ColumnTransformer to set the estimators and transformations

preprocessing = ColumnTransformer([('num', num_pipeline, num_cols),
                                   ('cat', cat_pipeline, cat_cols)],
                                    remainder='passthrough'
                                 )
```

```python
num_cols
```

```python
cat_cols
```

```python
preprocessing
```

```python
X_prepared = preprocessing.fit_transform(X)

# add column headers back.
feature_names=preprocessing.get_feature_names_out()

# Create a new DataFrame from the transformed data 'X_prepared'.
# Assign the feature names to the columns of this new DataFrame
X_prepared = pd.DataFrame(data=X_prepared, columns=feature_names)

X_prepared
```

<h2>Now split the test set.<h2>

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_prepared, Y, test_size=0.2, random_state=42)
# Check if the dataset is balanced or not (The answer is yes, based on the shape)
print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

<h3>Scale down the target values(Y) since we are guessing millions value.<h3>

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
#Going to need scale down train and test set both.
y_train_scaled = scaler.fit_transform(y_train.values.reshape(-1, 1))
y_test_scaled = scaler.transform(y_test.values.reshape(-1, 1))
```

<h3>Trying random forest algorithms<h3>

```python
from sklearn.metrics import mean_squared_error as mse
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score


rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
#I used ravel() inorder to convert y variable to 1D array
rf_model.fit(X_train, y_train_scaled.ravel())
y_pred = rf_model.predict(X_test)
from sklearn.metrics import mean_squared_error as mse
#Calculating mean scored value.
RR_mse=mse(y_test_scaled, y_pred)
#ACCURACY AS A PERCENTAGES
print(f'Mean Squared Error: {RR_mse}')
```

<h2>Now try linear regression algorithms.<h2>

```python
from sklearn.linear_model import LinearRegression

lr_model = LinearRegression()

lr_model.fit(X_train,y_train_scaled)

lr_y_predict = lr_model.predict(X_test)
#MSE for linear regression.
lr_mse=mse(y_test_scaled, lr_y_predict)
print(f'Mean Squared Error: {lr_mse}')
```

<h3>Checking validation score with 5 folds to see how well with unseen data, the average of it.(For linear regression)<h3>

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(lr_model, X_train, y_train_scaled, cv=5, scoring='neg_mean_squared_error')

mean_score = -scores.mean()  # Take the negative value to get the mean squared error

print(f'Cross-Validation Mean Score: {mean_score}')
```

<h2>Now calculate the cross validation with 5 folds for Randomforest.<h2>

```python

scores = cross_val_score(rf_model, X_train, y_train_scaled.ravel(), cv=5, scoring='neg_mean_squared_error')

mean_score = -scores.mean()  # Take the negative value to get the mean squared error

print(f'Cross-Validation Mean Score: {mean_score}')
```

<h3>Now do a regularization(Ridge, Lasso) to eaze the weight on the graph.(linear regression).<h3>

```python
from sklearn.linear_model import Ridge, Lasso

RidgeRegression = Ridge(alpha=1)
ridge_model = RidgeRegression.fit(X_train, y_train_scaled)


LassoRegression = Lasso(alpha=1)
lasso_model = LassoRegression.fit(X_train, y_train_scaled)
```

<h2>Calculate MSE for regularization.<h2>


```python
Ridge_y_predict = ridge_model.predict(X_test)
Lasso_y_predict = lasso_model.predict(X_test)
ridge_mse = mse(y_test_scaled, Ridge_y_predict)
lasso_mse=mse(y_test_scaled, Lasso_y_predict)

print(f'Ridge Regression MSE: {ridge_mse}')
print(f'Lasso Regression MSE: {lasso_mse}')
```
